```python
#Import Libraries
from bs4 import BeautifulSoup 
import requests
import re
import csv
import pandas
```

```python
#Grab data from url and parse into HTML elements using BS
r = requests.get('https://www.okhistory.org/research/1890results.php?pageno=10&lname=%&fname=%&action=Search')
soup=BeautifulSoup(r.text, 'html.parser')
```

```python
num = 1
names = []
ages =[]
genders = []
races = []
birthplaces = []
pages = []

while num <= 1149:
    print('Page Number:',str(num))
    r = requests.get('https://www.okhistory.org/research/1890results.php?pageno='+
                    str(num)+'&lname=%&fname=%&action=Search')
    soup=BeautifulSoup(r.text, 'html.parser')
    tables = soup.findAll('td')
    length = len(tables)
    print(length)
    n=0
    while n < length:
        print('name:',tables[n].getText())
        names.append(tables[n].getText())
        ages.append(tables[n+1].getText())
        genders.append(tables[n+2].getText())
        races.append(tables[n+3].getText())
        birthplaces.append(tables[n+4].getText())

        pages.append(tables[n+5].getText())
        n= n+6
    num = num+1
```

```python
data = pandas.DataFrame({'Name':names,
                        'Age':ages,
                        'Gender':genders,
                        'Race':races,
                        'Birthplace':birthplaces,
                        'Page':pages})
```
